Filtering a Data Frame
-----------------------

```python
import pandas as pd
pd.read_csv('netflix_titles.csv')
```

```python
netflix_titles = pd.read_csv('netflix_titles.csv', parse_dates=['date_added'])
netflix_titles.info()
```

```python
netflix_titles["country"].fillna('United States').tail(5)
netflix_titles.nunique()
```

```python
# lets categorize the data
netflix_titles["type"].astype('category')
```

```python
# filter by release year (Single column)

netflix_titles[netflix_titles['release_year'] == 2021]
```

```python
# filter by multiple conditions
# And

is_movie = netflix_titles['type'] == "Movie"
is_recent_release = netflix_titles['release_year'] > 2019

# all the movies released post 2019
netflix_titles[is_movie & is_recent_release]
```

```python
# OR condition
is_movie = netflix_titles['type'] == "Movie"
is_indian = netflix_titles['country'] == "India"

netflix_titles[is_movie | is_indian]


```

```python
# inversion with ~

my_series = pd.Series([True, False, True])
print(my_series)
print("inverted series")
print(~my_series)

# lets apply this 

is_indian = netflix_titles['country'] == "India"
netflix_titles[~is_indian]
```

```python
# isin

is_american = netflix_titles['country'] == "United States"
is_indian = netflix_titles['country'] == "India"

print(netflix_titles[is_american | is_indian])

print('############################')
countries = ['India', 'United States']
is_title_in_countries = netflix_titles['country'].isin(countries)
netflix_titles[is_title_in_countries]
```

```python
# between
# write an expression to find all the titles between 1990 and 2000

between_1990_2000 = netflix_titles['release_year'].between(left=1990, right=2000)
netflix_titles[between_1990_2000]
```

```python
# Find the records where the country,cast and director are not null
is_country_null = netflix_titles['country'].isnull()
is_cast_null = netflix_titles['cast'].isnull()
is_director_null = netflix_titles['director'].isnull()
netflix_titles[~(is_country_null  | is_cast_null | is_director_null)]
```

```python
# Dealing with Duplicates

is_duplicated = netflix_titles["country"].duplicated()
netflix_titles[is_duplicated]

# drop duplicates in listed_in
netflix_titles.drop_duplicates(subset=["listed_in"], keep='first')

# drop duplicates in date_added
netflix_titles.drop_duplicates(subset=["date_added"], keep='last')
```

```python
import pandas as pd
delays_percentage = {'Carrier': [23.02], 'ATC': [30.42], 'Weather': [4.03], 'Security': [0.12], 'Inbound': [42.43]}
dfw = pd.DataFrame.from_dict(delays_percentage)
```

```python
ax = dfw.transpose().plot.bar(legend=False)
ax.set_xlabel('Cause of delay')
ax.set_ylabel('Count')
```

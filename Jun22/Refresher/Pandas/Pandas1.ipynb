```python
by_india = shows[shows["country"]=="India"]
movies_by_india = by_india[by_india["type"] == "Movie"]
movies_by_india
```

```python
shows_by_india = shows["country"] == "India"
movies = shows["type"] == "Movie"
shows[shows_by_india & movies].head(3)

# indian movies before 2000
before_2000 = shows["release_year"] <= 2000
shows_by_india = shows["country"] == "India"
movies = shows["type"] == "Movie"
shows[before_2000 & shows_by_india & movies].head(2)
```

```python
# movies release between 2018 and 2021
between_2018_n_2021 = shows["release_year"].between(2018,2021)
movies = shows["type"] == "Movie"

shows[between_2018_n_2021 & movies]
```

```python
# Remove min from duration
(shows["duration"]
    .dropna()
    .str.replace(" min","", regex=False)
    .str.replace(" Seasons","", regex=False)
    .str.replace(" Season","", regex=False)
    .astype(int)
)
```
